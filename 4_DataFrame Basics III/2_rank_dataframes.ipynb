```python
import pandas as pd
```

```python
sales = pd.Series([15,30,45,21,55,15,0], index = ['Mon','Tue','Wed','Thurs','Fri','Sat','Sun'], name = 'Sales')
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank()
```

```python
sales.rank(ascending = False, method = 'min', pct = False).sort_values(ascending = True)
```

```python
sales.rank(ascending = False, method = 'min', pct = True).sort_values(ascending = True)
```
