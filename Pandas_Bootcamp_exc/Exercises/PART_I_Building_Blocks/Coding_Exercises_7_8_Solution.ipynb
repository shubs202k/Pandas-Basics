61. __Calculate__ for all __numerical columns__ the __range between highest and lowest__ value! __Fill in the gaps!__ The range for __mpg__ is...?

```python
cars.iloc[:, :-2].apply(lambda x: x.max() - x.min(), axis = 0)
```
