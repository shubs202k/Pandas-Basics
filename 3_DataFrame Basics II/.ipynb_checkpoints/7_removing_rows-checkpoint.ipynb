## Drop rows that do not meet a certain condition

```python
summer = pd.read_csv('summer.csv', index_col = 'Athlete')
```

```python
mask1 = summer.Year == 1996
mask2 = summer.Sport == 'Aquatics'
```

```python
summer = summer.loc[~(mask1 | mask2)]
```

```python
summer.head()
```
