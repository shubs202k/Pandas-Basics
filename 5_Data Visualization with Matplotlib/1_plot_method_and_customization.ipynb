## Customization

```python
titanic.age.plot(figsize = (12,8), fontsize = 15, c = 'black', linestyle = '-', 
                 xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks,
                 rot = 45) #rot = 90 rotate xticks by 90 degrees
plt.title('Titanic-Ages', fontsize = 18)
plt.legend(loc = 'best', fontsize = 14)
plt.xlabel('Passenger No', fontsize = 14)
plt.ylabel('Age', fontsize = 14)
#plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0,901,50)]
yticks = [y for y in range(0,81,5)]
```
