## Scatterplots : Useful in detecting relationships between numerical features

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic.plot(x = 'age', y ='fare', kind = 'scatter', figsize = (15,8), c = 'red', marker = 'x', s = 50)
# s:size of the scatter
plt.show()
```

## 3-D Scatterplot

```python
titanic.plot(x = 'age', y ='fare', kind = 'scatter', figsize = (15,8), c = 'survived', 
             marker = 'x', s = 50, colormap = 'viridis')
# s:size of the scatter
plt.show()
```

```python
titanic.plot(x = 'age', y ='fare', kind = 'scatter', figsize = (15,8), c = 'pclass', 
             marker = 'x', s = 50, colormap = 'viridis')
# s:size of the scatter
plt.show()
```
