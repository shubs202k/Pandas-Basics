## Select rows with iloc (Position based indexing)

```python
summer = pd.read_csv('summer.csv')
```

```python
summer.head()
```

```python
summer.tail()
```

```python
summer.info()
```

```python
summer.iloc[0] #first row
```

```python
type(summer.iloc[0])
```

```python
summer.iloc[1] #second row
```

```python
summer.iloc[-1] #last row
```

```python
summer.iloc[:5] #first five rows
```

```python
summer.head()
```

```python
summer.iloc[-5:] #last five rows
```

```python
summer.iloc[500:511]
```

```python
#get rows from different indexes
summer.iloc[[1,1000,6000,15000,20000,30000]]
```
