## Pandas by default opens the first sheet in the excel file

```python
pd.read_excel('summer_raw.xls', sheet_name = 'summer', skiprows = 2)
```

```python
pd.read_excel('summer_raw.xls', sheet_name = 'summer', skiprows = [0,1])
```

```python
pd.read_excel('summer_raw.xls', sheet_name = 'summer', skiprows = [0,1], usecols = 'C:L')
```

```python
summer = pd.read_excel('summer_raw.xls', sheet_name = 'summer', skiprows = [0,1], usecols = 'D:L')
```
