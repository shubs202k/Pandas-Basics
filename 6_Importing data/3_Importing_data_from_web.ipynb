## Importing data from web using pd.read_html()

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/1976_Summer_Olympics_medal_table'
```

```python
pd.read_html(url)
```

```python
pd.read_html(url)[1]
```

```python
wiki_1976 = pd.read_html(url)[0]
```
